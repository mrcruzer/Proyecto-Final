```python
#Importamos el json y expotamos a csv
for i in range(1,50):
    try:
        alabama = pd.read_json('reviews-estados/review-Alabama/'+str(i)+'.json',lines=True)     #Importamos el json
        alabama.drop(columns=['pics','resp'],inplace=True)              #Dropeamos las columnas que no vamos a usar
        for j in alabama.dropna(subset='text')[alabama.dropna(subset='text')['text'].str.contains('\n')].index:
            alabama['text'][j] = alabama['text'][j].replace('\n','')    #Quitamos estos caracteres raros
        alabama.to_csv('reviews-estados/review-Alabama/'+str(i)+'.csv',index=False,escapechar='\\',sep=';')     #Exportamos a csv
    except:
        print(str(i))
        break
```

```python
#Funcionó bien. Definimos función y la corremos con los otros datasets.
def jsonAcsv (lugar:str):
    for i in range(1,50):
        try:
            lugar_df = pd.read_json('reviews-estados/review-'+lugar+'/'+str(i)+'.json',lines=True)     #Importamos el json
            lugar_df.drop(columns=['pics','resp'],inplace=True)              #Dropeamos las columnas que no vamos a usar
            for j in lugar_df.dropna(subset='text')[lugar_df.dropna(subset='text')['text'].str.contains('\n')].index:
                lugar_df['text'][j] = lugar_df['text'][j].replace('\n','')    #Quitamos estos caracteres raros
            lugar_df.to_csv('reviews-estados/review-'+lugar+'/'+str(i)+'.csv',index=False,escapechar='\\',sep=';')     #Exportamos a csv
        except:
            print(str(i))
            break
```
