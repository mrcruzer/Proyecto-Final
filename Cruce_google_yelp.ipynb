```python
#Importamos pandas
import pandas as pd
```

```python
#Importamos el dataset de google
google = pd.read_pickle('restaurants_google.pkl')
google.head()
```

```python
google.info()
```

```python
#Importamos el dataset de yelp
yelp = pd.read_pickle('restaurants_yelp.pkl')
yelp.head()
```

```python
yelp.info()
```

```python
#Reducimos los dataframes así es más liviano el trabajo
google_id_nombre_postal = google[['Id_Restaurant','Nombre','Cod_postal']]
yelp_id_nombre_postal = yelp[['Id_Restaurant','Nombre','Cod_postal']]
```

```python
#Juntamos los dataframes
merge = pd.merge(google_id_nombre_postal,yelp_id_nombre_postal,on=['Nombre','Cod_postal'])
merge.head()
```

```python
#Hay 5154 registros que coinciden en los dos datasets
merge.info()
```
