```python
#Importamos librerias
import pandas as pd
import numpy as np
```

```python
#Importamos dataset de negocios de yelp
yelp_restaurants = pd.read_pickle('restaurants_yelp.pkl')
yelp_restaurants.head()
```

```python
#Observamos mas de cerca
yelp_restaurants.info()
```

Este caso parece ser relativamente más sencillo. Prácticamente no hay valores nulos. Simplemente hay que descartar columnas, renombrar y reordenar.

```python
yelp_restaurants.drop(columns=['index','address','stars','review_count','is_open','hours'],inplace=True)
yelp_restaurants.info()
```

```python
#Renombramos
yelp_restaurants.rename(columns={'business_id':'Id_Restaurant','name':'Nombre','city':'Ciudad','state':'Estado','postal_code':'Cod_postal',
                                 'latitude':'Latitud','longitude':'Longitud','attributes':'Atributos','categories':'Tipo'},inplace=True)
yelp_restaurants.head()
```

```python
#Reordenamos
yelp_restaurants = yelp_restaurants[['Id_Restaurant', 'Nombre', 'Ciudad', 'Estado', 'Cod_postal', 'Latitud',
                                    'Longitud','Tipo','Atributos']]
yelp_restaurants.head()
```

```python
#Exportamos
yelp_restaurants.to_pickle('restaurants_yelp.pkl')
```
