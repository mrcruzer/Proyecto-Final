```python
#Importamos librerias
import pandas as pd
```

```python
#Importamos dataset de negocios de yelp
yelp_restaurants = pd.read_pickle('restaurants_yelp.pkl')
yelp_restaurants.head()
```

```python
#Observamos mas de cerca
yelp_restaurants.info()
```

Este caso parece ser relativamente más sencillo. Prácticamente no hay valores nulos. Simplemente hay que descartar columnas, renombrar y reordenar.

```python
yelp_restaurants.drop(columns=['index','address','stars','review_count','is_open','hours'],inplace=True)
yelp_restaurants.info()
```

```python
#Renombramos
yelp_restaurants.rename(columns={'business_id':'Id_Restaurant','name':'Nombre','city':'Ciudad','state':'Estado','postal_code':'Cod_postal',
                                 'latitude':'Latitud','longitude':'Longitud','attributes':'Atributos','categories':'Tipo'},inplace=True)
yelp_restaurants.head()
```

```python
#Reordenamos
yelp_restaurants = yelp_restaurants[['Id_Restaurant', 'Nombre', 'Ciudad', 'Estado', 'Cod_postal', 'Latitud',
                                    'Longitud','Tipo','Atributos']]
yelp_restaurants.head()
```

```python
yelp_restaurants.info()
```

```python
#Vamos a observar si hay negocios que se encuentren duplicados.
#Es decir que tengan distinto Id (ya se descartó antes los de igual Id) pero coincidan nombre y ubicación.
duplicados_yelp = yelp_restaurants[yelp_restaurants.duplicated(subset=['Nombre','Cod_postal'],keep=False)].sort_values(by='Nombre')
duplicados_yelp.head()
```

```python
#Vemos que hay y tienen la misma ubicación. Vemos cuantos son.
duplicados_yelp.reset_index(inplace=True)
duplicados_yelp.info()
```

Es un numero impar. Es decir que hay algunos que están repetidos mas de 1 vez.

```python
#Vemos cuantos negocios son
negocios_duplicados_yelp = duplicados_yelp.drop_duplicates(subset=['Nombre','Cod_postal'])
negocios_duplicados_yelp.info()
```

Vamos a quedarnos con uno de los duplicados y descartar los otros. Sin embargo, vamos a cambiar primero el id de las reseñas así no perdemos información.

```python
#Tomamos los ids a conservar y los que se van a cambiar
lista_duplicados = []
for n in negocios_duplicados_yelp['Id_Restaurant'].values:
    aux = []
    nombre = negocios_duplicados_yelp[negocios_duplicados_yelp['Id_Restaurant']==n]['Nombre'].values[0]
    cod_postal = negocios_duplicados_yelp[negocios_duplicados_yelp['Id_Restaurant']==n]['Cod_postal'].values[0]
    for d in range(duplicados_yelp.shape[0]):
        if (duplicados_yelp['Nombre'][d] == nombre) and (duplicados_yelp['Cod_postal'][d] == cod_postal) and (duplicados_yelp['Id_Restaurant'][d] != n):
            aux.append(duplicados_yelp['Id_Restaurant'][d])
    lista_duplicados.append([n,aux])
```

```python
#Armammos dataframe
duplicados_yelp_df = pd.DataFrame(data=lista_duplicados,columns=['Conservar','Cambiar'])
duplicados_yelp_df.head()
```

```python
#Descartamos los duplicados
yelp_restaurants.drop_duplicates(subset=['Nombre','Cod_postal'],inplace=True)
yelp_restaurants.reset_index(inplace=True)
yelp_restaurants.drop(columns='index',inplace=True)
yelp_restaurants.info()
```

```python
#Exportamos
yelp_restaurants.to_pickle('restaurants_yelp_final.pkl')
```
