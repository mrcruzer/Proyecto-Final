```python
#Importamos pandas
import pandas as pd
```

```python
#Importamos listado de negocios de yelp
business_yelp = pd.read_pickle('business.pkl')
business_yelp.head()
```

```python
#Observamos mas de cerca
business_yelp.info()
```

Vemos que se repiten las columnas y que el segundo set de campos solo tiene 5 valores no nulos. Vamos a observar con mayor detenimiento estos así nos aseguramos que al descartar no se pierda informacion.

```python
#Vemos los registros que tienen info en el segundo set de columnas que se repiten
business_yelp[business_yelp.index.isin(business_yelp.iloc[:,14:].dropna().index)]
```

```python
#Los datos son los mismos entonces tomamos las primeras 13 columnas.
business_yelp_filt = business_yelp.iloc[:,:14]
business_yelp_filt.head()
```

```python
#Tomamos solo los que tengan restaurant bajo category. Dropeamos primero entonces las categories nulas
business_yelp_filt.dropna(subset='categories',inplace=True)
business_yelp_filt.info()
```

```python
#Filtramos unicamente los restaurants
restaurants_yelp = business_yelp_filt[business_yelp_filt['categories'].str.contains('estaura')]
restaurants_yelp.head()
```

```python
#Reseteamos el índice y descartamos los valores duplicados de business_id
restaurants_yelp.reset_index(inplace=True)
restaurants_yelp.drop_duplicates(subset='business_id',inplace=True)
restaurants_yelp.info()
```

```python
#Exportamos el archivo
restaurants_yelp.to_pickle('restaurants_yelp.pkl')
```

Transformación del archivo de las reseñas de Yelp (review.json)

```python
#Haremos la importación y al mismo tiempo, filtraremos para quedarnos solo con los que coincidan con el campo business_id
for i in range (1,15):
    df =pd.read_json(f"Yelp\{i}.json", lines=True)
    df_filtrado = df[df["business_id"].isin(restaurants_yelp["business_id"].values)]
    df_filtrado.to_csv(fr"Yelp\{i}.csv", index=False)
```
