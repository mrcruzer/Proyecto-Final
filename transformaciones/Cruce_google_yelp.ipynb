```python
#Importamos pandas
import pandas as pd
import numpy as np
```

```python
#Importamos el dataset de google
google = pd.read_pickle('metadata-sitios/restaurants_google.pkl')
google.head()
```

```python
google.info()
```

```python
#Importamos el dataset de yelp
yelp = pd.read_pickle('yelp/restaurants_yelp.pkl')
yelp.head()
```

```python
yelp.info()
```

```python
#Reducimos los dataframes así es más liviano el trabajo
google_id_nombre_postal = google[['Id_Restaurant','Nombre','Cod_postal']]
yelp_id_nombre_postal = yelp[['Id_Restaurant','Nombre','Cod_postal']]
```

```python
#Juntamos los dataframes
merge = pd.merge(google_id_nombre_postal,yelp_id_nombre_postal,on=['Nombre','Cod_postal'])
merge.head()
```

```python
#Hay 3359 registros que coinciden en los dos datasets
merge.info()
```

```python
#Observamos los valores en el dataframe de google
google[google['Id_Restaurant'].isin(merge['Id_Restaurant_x'].values)].sort_values(by='Nombre')
```

```python
#Observamos los valores en el dataframe de yelp
yelp[yelp['Id_Restaurant'].isin(merge['Id_Restaurant_y'].values)].sort_values(by='Nombre')
```

Vamos a conservar los ids de google y descartar los de yelp. Entonces deberemos reemplazar los segundos con los primeros en las reviews.

```python
#Descartamos los registros del dataset de yelp
indices_descarte = yelp[yelp['Id_Restaurant'].isin(merge['Id_Restaurant_y'].values)].index
yelp.drop(index=indices_descarte,inplace=True)
yelp.reset_index(inplace=True,drop=True)
yelp.info()
```

```python
#Armamos un solo dataframe de negocios
restaurants = pd.concat([google,yelp],ignore_index=True)
restaurants.head()
```

```python
restaurants.info()
```

```python
#Exportamos
restaurants.to_pickle('restaurants.pkl')
```

Ahora resta modificar las reviews para que todas las que eliminamos de yelp pasen a referirse a google.

```python
#Importamos las reviews de google
reviews_google = pd.read_csv('reviews-estados/reviews.csv',sep=';',escapechar='\\')
reviews_google.info()
```

```python
#Importamos las reviews de yelp
reviews_yelp = pd.read_csv('yelp/reviews_yelp.csv',sep=';',escapechar='\\')
reviews_yelp.info()
```

```python
#Tomamos un valor del dataframe merge para observarlo en el de las reviews de yelp para chequear luego el cambio
reviews_yelp[reviews_yelp['Id_Restaurant']=='g2bYK-Z3foqSnCiOPYtdsA']
```

```python
merge.head()
```

```python
#Le quitamos las columnas de Nombre y código postal a merge para que sea mas liviano
merge.drop(columns=['Nombre','Cod_postal'],inplace=True)
```

```python
#Lo mergeamos con el de review de yelp
reviews_yelp_merge = pd.merge(reviews_yelp,merge,how='left',left_on='Id_Restaurant',right_on='Id_Restaurant_y')
reviews_yelp_merge.head()
```

```python
#Observamos algun valor presente
reviews_yelp_merge[reviews_yelp_merge['Id_Restaurant']=='g2bYK-Z3foqSnCiOPYtdsA']
```

```python
#Hacemos unos cambios de nombre para poder llenar los nulos de Id_Restaurant_x
reviews_yelp_merge.rename(columns={'Id_Restaurant':'Id_Restaurant_orig'},inplace=True)
reviews_yelp_merge.rename(columns={'Id_Restaurant_x':'Id_Restaurant'},inplace=True)
reviews_yelp_merge.head()
```

```python
#Lo llenamos con los valores del dataframe de reviews de yelp
reviews_yelp_merge.fillna(reviews_yelp,inplace=True)
reviews_yelp_merge.head()
```

```python
merge.head()
```

```python
#Chequeamos algunos valores que se cambiaron
reviews_yelp_merge[reviews_yelp_merge['Id_Restaurant_orig']=='H1bbYNKgk6JF9pKcBXyDXw'].head()
```

```python
#Anduvo perfecto. Modificamos una última vez
reviews_yelp = reviews_yelp_merge[['Id_Usuario','Id_Restaurant','Rating','Reseña','Fecha']]
reviews_yelp.head()
```

```python
#Cambiamos el formato de la fecha a timestamp así coincide con el de las reviews de google
reviews_yelp['Timestamp'] = pd.to_datetime(reviews_yelp['Fecha']).astype(np.int64)
reviews_yelp.head()
```

```python
reviews_yelp.drop(columns='Fecha',inplace=True)
reviews_yelp.info()
```

```python
reviews_google.info()
```

```python
#Armamos un único dataframe de reviews
reviews = pd.concat([reviews_google,reviews_yelp],ignore_index=True)
reviews.info()
```

```python
reviews['Id_Usuario'].unique().size
```

```python
reviews['Id_Restaurant'].unique().size
```

```python
#Exportamos en 8 datasets
N = 2000000
total_exp = 0
parte = 1
while total_exp < reviews.shape[0]:
    try:
        reviews.iloc[total_exp:total_exp+N].to_csv('reviews_finales/reviews_'+str(parte)+'.csv',index=False,sep=';',escapechar='\\')
        parte += 1
        total_exp += N
    except:
        reviews.iloc[total_exp:].to_csv('reviews_finales/reviews_'+str(parte)+'.csv',index=False,sep=';',escapechar='\\')
        total_exp += N
```
