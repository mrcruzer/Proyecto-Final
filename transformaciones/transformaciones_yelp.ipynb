```python
#Vamos a observar si hay negocios que se encuentren duplicados.
#Es decir que tengan distinto Id (ya se descartÃ³ antes los de igual Id) pero coincidan nombre y ubicaciÃ³n.
duplicados_yelp = restaurants_yelp[restaurants_yelp.duplicated(subset=['Nombre','Cod_postal'],keep=False)].sort_values(by='Nombre')
duplicados_yelp.head()
```

```python
#Vemos que hay y tienen la misma ubicaciÃ³n. Vemos cuantos son.
duplicados_yelp.reset_index(inplace=True)
duplicados_yelp.info()
```

```python
#Vemos cuantos negocios son
negocios_duplicados_yelp = duplicados_yelp.drop_duplicates(subset=['Nombre','Cod_postal'])
negocios_duplicados_yelp.info()
```

Vamos a quedarnos con uno de los duplicados y descartar los otros. Sin embargo, vamos a cambiar primero el id de las reseÃ±as asÃ­ no perdemos informaciÃ³n.

```python
#Tomamos los ids a conservar y los que se van a cambiar
lista_duplicados = []
for n in negocios_duplicados_yelp['Id_Restaurant'].values:
    aux = []
    nombre = negocios_duplicados_yelp[negocios_duplicados_yelp['Id_Restaurant']==n]['Nombre'].values[0]
    cod_postal = negocios_duplicados_yelp[negocios_duplicados_yelp['Id_Restaurant']==n]['Cod_postal'].values[0]
    for d in range(duplicados_yelp.shape[0]):
        if (duplicados_yelp['Nombre'][d] == nombre) and (duplicados_yelp['Cod_postal'][d] == cod_postal) and (duplicados_yelp['Id_Restaurant'][d] != n):
            aux.append(duplicados_yelp['Id_Restaurant'][d])
    lista_duplicados.append([n,aux])
```

```python
#Armammos dataframe
duplicados_yelp_df = pd.DataFrame(data=lista_duplicados,columns=['Conservar','Cambiar'])
duplicados_yelp_df.head()
```

```python
#Juntamos todos los ids que hay que modificar
ids_modificar = []
for i in range(duplicados_yelp_df.shape[0]):
    ids_modificar+=duplicados_yelp_df['Cambiar'][i]
ids_modificar
```

```python
#Los vemos en el dataframe de los restaurants
restaurants_yelp[restaurants_yelp['Id_Restaurant'].isin(ids_modificar)]
```

```python
#Descartamos los duplicados
restaurants_yelp.drop(index=restaurants_yelp[restaurants_yelp['Id_Restaurant'].isin(ids_modificar)].index,inplace=True)
restaurants_yelp.reset_index(inplace=True,drop=True)
restaurants_yelp.info()
```

```python
# exportamos la data hacia Restaurantes

for index, row in restaurants_yelp.iterrows():
        cursor.execute("INSERT INTO dbo.Restaurantes (Id_Restaurante, Nombre, Ciudad, Estado, Cod_postal, Latitud, Longitud, Tipo, Atributos) values(?, ?, ?, ?, ?, ?, ?, ?, ?)", 
                  row.Id_Restaurante, row.Nombre, row.Ciudad, row.Estado, row.Cod_postal, row.Latitud, row.Longitud, row.Tipo, row.Atributos
                   )
    
    
cursor.commit()

print("Termine de ingresar datos")
```

```python
#Definimos la funciÃ³n para realizar el reemplazo de los ids de los restaurants ya probada con el dataset de google
def cambioDeIndices (estado_df):
    for f in range(duplicados_yelp_df.shape[0]):
        reemplazador = duplicados_yelp_df['Conservar'][f]
        for c in duplicados_yelp_df['Cambiar'][f]:
            if estado_df[estado_df['Id_Restaurant']== c].index.size>0:
                for i in estado_df[estado_df['Id_Restaurant']== c].index:
                    estado_df.at[i,'Id_Restaurant'] = reemplazador
    return estado_df
```
